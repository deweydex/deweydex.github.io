- Takes into two integers

- Return the result.

2. Function ``numbersquish()`` that:
- overloads the previous method
- takes in an Array of integers
- concatonates the Array of integers all together
- returning the result.

```
public string numbersquish(int a, int b)
{
    string numStr1 = a.ToString();
    string numStr2 = b.ToString();
    return numStr1+" "+numStr2;
}
public string numbersquish(int[] intArray)
{
    string intStr = "";
    for(int i=0; i<intArray.Length; i++){
        intStr += intArray[i].ToString();
    }
    return intStr;
}

Console.WriteLine(numbersquish(1,1));
Console.WriteLine("---------------------");
int[] myArray = {34,2,56,27};
Console.WriteLine(numbersquish(myArray));
```


### Question C1
Please complete the following:

-  Create a method `DisplayGuide()` that takes in a list/array of Planets as input, and prints the name of each planet along with its safety status based upon the ``isDangerous()`` function. 

(e.g., "Earth - Safe", "Planet X - Dangerous").


- Create a list or array of `Planet` Objects called `GalacticGuide` containing the 3 Locations provided (``planetX``,``trillian``,``vogsphere``).

```
// Write your DisplayGuide() method 

public void DisplayGuide(List<Planet> Planets){
 
    foreach(var planet in Planets){
        if(planet.isDangerous()){
            Console.WriteLine($"Planet {planet.Name} is Dangerous!");
        }else{
            Console.WriteLine($"Planet {planet.Name} is Safe!");
        } 
    }
}


// here are some sample planets for your GalacticGuide
Planet planetX = new Planet("Beti Alpha", 7 );
Planet trillian = new Planet("Trillian's Home", 2);
Planet vogsphere = new Planet("Vogsphere", 10);

// declare your list/array of Planets named GalacticGuide here

List<Planet> GalacticGuide = new List<Planet>{planetX,trillian,vogsphere};
// uncomment the code below to test your function and GalacticGuide

DisplayGuide(GalacticGuide);
```

### Question C2

Please compete the following: 
- Create a method `GetSafeLocations()` that takes in a list/array of ``Planet`` Objects returns a integer count of `Planet` objects that are marked as safe based upon the ``isDangerous()`` function.
- Create a list or array of `Location` Objects called `GalacticGuide` containing the 6 Locations provided (``Earth``, ``Magrathea``, ``Mars``, ``planetX``,``trillian``,``vogsphere``).

```

// Write your GetSafeLocations() method here

public int GetSafeLocations(Planet[] planets){
    
    int num_safe_planets = 0;
    foreach(var planet in planets){
        
        if(planet.isDangerous() == false){
            num_safe_planets++;
        }
        
    }
    return num_safe_planets;
}


// here are some sample planets for your GalacticGuide
Planet earth = new Planet("Earth", 2);
Planet magrathea = new Planet("Magrathea", 9);
Planet mars = new Planet("Mars", 4);
Planet planetX = new Planet("Beti Alpha", 7 );
Planet trillian = new Planet("Trillian's Home", 2);
Planet vogsphere = new Planet("Vogsphere", 10);

// declare your list/array of Planets named GalacticGuide here
Planet[] GalacticGuide = new Planet[]{earth,magrathea,mars,planetX,trillian,vogsphere};

// uncomment the code below to test your function and GalacticGuide

int safe_planets = GetSafeLocations(GalacticGuide);
Console.WriteLine($"there is/are {safe_planets} safe planet(s) in our guide!");
```
