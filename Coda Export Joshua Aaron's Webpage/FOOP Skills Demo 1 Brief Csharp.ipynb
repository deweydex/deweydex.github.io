# Skills Demo 1 Brief: Classless Store 

In this skills demo you will be tasked with creating a shop and maintaining a personal inventory. Here, we will outline a few functions and requirements for your shop, the items in it, and user interaction (and inventory maintenance). You are welcome to complete this assignment in any modern programming language, though of course we have focused on Python or C#. If you would like to use another programming language from those two, please speak to me so we can align on questions of code style and structure.

This demo will be broken into 5 tasks, but you are welcome to (and it is in fact recommented that you) include other helper functions and global variables.

## Tasks
1. Describe a series of items with associated costs, names, and quantities, establish how to store these in a list, series of lists, or list like structure. (5%)

2. Define a series of variables for: (5%)
    - storing your users inventory
    - some means of purchasing or trading for new items
    - a means of storing your shops inventory 

3. Design two functions that display what is available at your shop and what is in the users inventory. (10%)

4. Design a function that allows for the purchasing of an item: (10%)
- make sure to check if appropriate trade or currency is available from the user
- ensure the inventory of your shop and user are updated appropriately after purchase

5. Introduce your user to your shop and provide a means for them to purchase another item, allow the user to exit the shop. (5%)
